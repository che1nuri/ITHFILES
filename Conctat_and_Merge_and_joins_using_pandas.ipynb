```python
df1= mydata[mydata['original_language']=='en']
df1
```

```python
 #Next let us create another dataframe which consists of only those movies whose length is greater than 100 mins
df2 = mydata.loc[mydata['runtime'] >= 100,['imdb_id','original_language']]
df2
```

Inner join 

```python
df3=pd.merge(df1,df2,how ='inner',left_on='imdb_id',right_on='imdb_id')
df3.shape
```

left join

```python
df4=pd.merge(df1,df2,how ='left',left_on='imdb_id',right_on='imdb_id')
df4.shape
```

right join

```python
df5=pd.merge(df1,df2,how ='right',left_on='imdb_id',right_on='imdb_id')
df5.shape
```

```python
merged_OJ_df = pd.merge(df1[['imdb_id','runtime']],df2, on='imdb_id', how='outer')
merged_OJ_df
```
